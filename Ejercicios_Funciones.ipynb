#Problema 1: Juego de "Piedra, Papel o Tijera"
Diseña un programa que permita al usuario jugar "Piedra, Papel o Tijera" contra la computadora. El
programa debe seguir estos pasos:
1. Al iniciar, se debe generar un número aleatorio entre 1 y 3:
o 1 significa que la computadora eligió "Piedra".
o 2 significa que la computadora eligió "Papel".
o 3 significa que la computadora eligió "Tijera". (No se muestra la elección de la
computadora todavía).
2. El usuario debe ingresar su elección: "Piedra", "Papel" o "Tijera".
3. Luego, se muestra la elección de la computadora.
4. El programa debe indicar quién es el ganador, siguiendo estas reglas:
o Piedra vence a Tijera (la rompe).
o Tijera vence a Papel (lo corta).
o Papel vence a Piedra (lo envuelve).
o Si ambos jugadores eligen lo mismo, el juego debe repetirse hasta que haya un
ganador.
5. Agrega un sistema de puntaje. El programa debe llevar un registro de cuántas veces gana el
usuario y cuántas veces gana la computadora.
6. Al finalizar cada ronda, se muestra el puntaje actual y se pregunta al usuario si desea seguir
jugando. Si el usuario responde "sí", el juego continúa. Si responde "no", el programa muestra
los puntajes finales.

START

```
import random

def jugar_tragamonedas():
    palabras = ["Cerezas", "Naranjas", "Ciruelas", "Campanas", "Melones", "Barras"]
    total_ingresado = 0
    total_ganado = 0

    while True:
        try:
            monto = float(input("¿Cuánto dinero desea ingresar a la máquina? $"))
            if monto <= 0:
                print("Por favor ingrese un monto mayor a 0.")
                continue
        except ValueError:
            print("Por favor ingrese un número válido.")
            continue

        total_ingresado += monto

        resultado = [random.choice(palabras) for _ in range(3)]
        print("Resultado:", resultado)

        if resultado[0] == resultado[1] == resultado[2]:
            ganancia = monto * 3
            print("¡Tres iguales! Usted gana $", ganancia)
        elif resultado[0] == resultado[1] or resultado[1] == resultado[2] or resultado[0] == resultado[2]:
            ganancia = monto * 2
            print("¡Dos iguales! Usted gana $", ganancia)
        else:
            ganancia = 0
            print("No hay coincidencias. Usted gana $0.")

        total_ganado += ganancia

        jugar_otra_vez = input("¿Desea jugar de nuevo? (sí/no): ").strip().lower()
        if jugar_otra_vez != "sí":
            break

    print("\nResumen del juego:")
    print(f"Total ingresado: ${total_ingresado}")
    print(f"Total ganado: ${total_ganado}")

# Ejecutar el juego
jugar_tragamonedas()
```

```
usuarios = {}

def crear_cuenta():
    print("\n--- Crear Cuenta ---")
    usuario = input("Ingrese nombre de usuario: ")
    if usuario in usuarios:
        print("Este nombre de usuario ya existe.")
        return
    contraseña = input("Ingrese contraseña: ")
    usuarios[usuario] = {
        "contraseña": contraseña,
        "saldo": 0,
        "historial": []
    }
    print("Cuenta creada exitosamente.")

def iniciar_sesion():
    print("\n--- Iniciar Sesión ---")
    usuario = input("Usuario: ")
    contraseña = input("Contraseña: ")
    if usuario in usuarios and usuarios[usuario]["contraseña"] == contraseña:
        print("Sesión iniciada.")
        menu_usuario(usuario)
    else:
        print("Usuario o contraseña incorrectos.")

def menu_usuario(usuario):
    while True:
        print("\n--- Menú Cajero ---")
        print("1. Depositar dinero")
        print("2. Consultar saldo")
        print("3. Retirar dinero")
        print("4. Ver historial de transacciones")
        print("5. Cerrar sesión")
        opcion = input("Seleccione una opción: ")

        if opcion == "1":
            depositar(usuario)
        elif opcion == "2":
            consultar_saldo(usuario)
        elif opcion == "3":
            retirar(usuario)
        elif opcion == "4":
            ver_historial(usuario)
        elif opcion == "5":
            print("Sesión cerrada.")
            break
        else:
            print("Opción no válida.")

def depositar(usuario):
    try:
        cantidad = float(input("Ingrese cantidad a depositar: "))
        if cantidad > 0:
            usuarios[usuario]["saldo"] += cantidad
            usuarios[usuario]["historial"].append(f"Depósito: +${cantidad:.2f}")
            print(f"Depósito exitoso. Nuevo saldo: ${usuarios[usuario]['saldo']:.2f}")
        else:
            print("Cantidad no válida.")
    except ValueError:
        print("Entrada inválida.")

def consultar_saldo(usuario):
    print(f"Saldo actual: ${usuarios[usuario]['saldo']:.2f}")

def retirar(usuario):
    try:
        cantidad = float(input("Ingrese cantidad a retirar: "))
        if 0 < cantidad <= usuarios[usuario]["saldo"]:
            usuarios[usuario]["saldo"] -= cantidad
            usuarios[usuario]["historial"].append(f"Retiro: -${cantidad:.2f}")
            print(f"Retiro exitoso. Nuevo saldo: ${usuarios[usuario]['saldo']:.2f}")
        else:
            print("Saldo insuficiente o cantidad inválida.")
    except ValueError:
        print("Entrada inválida.")

def ver_historial(usuario):
    print("\n--- Historial de Transacciones ---")
    historial = usuarios[usuario]["historial"]
    if historial:
        for transaccion in historial:
            print(transaccion)
    else:
        print("No hay transacciones registradas.")

def main():
    while True:
        print("\n--- Cajero Automático ---")
        print("1. Crear cuenta")
        print("2. Iniciar sesión")
        print("3. Salir")
        opcion = input("Seleccione una opción: ")

        if opcion == "1":
            crear_cuenta()
        elif opcion == "2":
            iniciar_sesion()
        elif opcion == "3":
            print("Gracias por usar el cajero. ¡Hasta luego!")
            break
        else:
            print("Opción no válida.")

if __name__ == "__main__":
    main()
```
